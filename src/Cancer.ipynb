__Logistic Regression__

```python
df.info()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:, 3:].values
y = df.iloc[:, 0].values
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lgr = LogisticRegression()
```

```python
lgr.fit(X_train, y_train)
```

```python
predY = lgr.predict(X_test)
```

```python
train_accuracy = lgr.score(X_train, y_train)
train_accuracy
```

```python
test_accuracy = lgr.score(X_test, y_test)
test_accuracy
```

```python
predY_series = pd.Series(predY)
```

```python
predY_series.value_counts().plot(kind='pie',labels=['B','M'],autopct='%1.1f%%')
plt.title("the ratio of predicted diagnosis")
plt.show()
```
